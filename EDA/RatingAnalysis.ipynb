```python
import pandas as pd
import numpy as np

df["resolution_outcome"] = pd.Categorical(df.resolution_outcome)

def locus_service(x):
    if x<5:
        return "negative"
    elif x>5:
        return "positive"
    return np.nan

df["service_grade_locus"] = df.service_grade.apply(locus_service)
df["service_grade_locus"] = pd.Categorical(df.service_grade_locus)

df.groupby(["resolution_outcome", "service_grade_locus"]).count()
```
